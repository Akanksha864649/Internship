```python
# Write s python program to display list of respected former presidents of India(i.e. Name , Term ofoffice) from https://presidentofindia.nic.in/former-presidents.htm and make data frame.

from bs4 import BeautifulSoup
import requests

page = requests.get('https://presidentofindia.nic.in/former-presidents.htm')

page

soup = BeautifulSoup(page.content)
soup

Name = soup.find('div',class_="presidentListing")
Name

Name = []
for i in soup.find_all('div',class_="presidentListing"):
    Name.append(i.text)
Name

Name = []

for i in soup.find_all("div",class_="presidentListing"): 

    Name.append(i.find("h3").text.split("(")[0])

Name

Term = soup.find("div",class_="presidentListing")
Term

Term = []

for i in soup.find_all("div",class_="presidentListing"): 

    Term.append(i.find('p').text.split("("))
Term

print(len(Name),len(Term))

import pandas as pd
df = pd.DataFrame({'Name':Name,'Term of Office':Term})
df

```

```python
# Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world and make data frame

from bs4 import BeautifulSoup
import requests

page = requests.get('https://www.cnbc.com/world/?region=world')

page

soup = BeautifulSoup(page.content)
soup

Headline = soup.find('a',class_="LatestNews-headline")
Headline

Headline.text

Headline = []

for i in soup.find_all('a',class_="LatestNews-headline"):
    Headline.append(i.text)
Headline

Time = []

for i in soup.find_all('time',class_="LatestNews-timestamp"):
    Time.append(i.text)
    
Time

URL = []

for i in soup.find_all('a',class_="LatestNews-headline"):
    URL.append(i.get('href'))
    
URL

print(len(Headline),len(Time),len(URL))

import pandas as pd
df = pd.DataFrame({'News Headline':Headline,'Time':Time,'News Link':URL})
df

```

```python
# Write a python program to scrape mentioned details from dineout.co.in and make data frame

from bs4 import BeautifulSoup
import requests

page = requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')

page

soup = BeautifulSoup(page.content)
soup

name = []

for i in soup.find_all('a',class_="restnt-name ellipsis"):
    name.append(i.text)
name

loc = []

for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    loc.append(i.text)
loc

cuisine = soup.find('span',class_="double-line-ellipsis")
cuisine.text.split('|')[1]

cuisine = []

for i in soup.find_all('span',class_="double-line-ellipsis"):
    cuisine.append(i.text.split('|')[1])
cuisine

rank = []

for i in soup.find_all('div',class_="restnt-rating rating-4"):
    rank.append(i.text)
rank

image = []

for i in soup.find_all('img',class_="no-img"):
    image.append(i.get('data-src'))
image

print(len(name),len(loc),len(cuisine),len(rank),len(image))

import pandas as pd
df = pd.DataFrame({'Restaurant Name':name,'Location':loc,'Cuisine':cuisine,'Ranking':rank,'Image':image})
df

```
